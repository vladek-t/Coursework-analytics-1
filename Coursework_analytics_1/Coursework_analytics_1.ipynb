```python
# Задание 1: (p) Рассчитать прирост студентов на каждом курсе в каждом месяце за всю историю 
# (каждый месяц в диапазоне от марта 2016 до июля 2019 включительно). Считать дату начала прохождения 
# курса студентом по дате начала первой домашней работы.

main_1 = pd.read_csv('main.csv', parse_dates=['start_date'], dayfirst=False, index_col=0)

df = df_course_content[df_course_content['is_homework'] == True].groupby('course_id').agg({'module_number' : 'min'}).reset_index()

df = df.rename(columns={'module_number' : 'first_module_with_homework'})
df['is_homework'] = True

main_1['year_month'] = main_1['start_date'].dt.to_period('M')
main_1

df = df.merge(main_1, left_on=['course_id', 'first_module_with_homework', 'is_homework'], right_on=['course_id', 'module_number', 'is_homework'], how='left')
df

df = df.groupby(['course_title', 'year_month', 'first_module_with_homework'])['student_id'].count().reset_index()
df = df.drop('first_module_with_homework', 1)
df
```

```python
# Задание 1: (p) Посчитать конверсию перехода студентов из одного модуля в другой на каждом курсе. Формула: отношение количества студентов, 
# приступивших к выполнению домашнего задания в этом модуле (если дз в модуле несколько, то считать по первому дз в модуле), к количеству 
# студентов, сдавших задание в предыдущем модуле (если дз в модуле несколько, то считать по последнему дз в модуле).

# Выборка основных данных
df = main.groupby(['course_title', 'module_number', 'lesson_number', 'is_homework', 'status']).agg('count').reset_index()[['course_title', 'module_number', 'lesson_number', 'is_homework', 'status', 'progress_id']]
df = df[df['is_homework'] == True].reset_index(drop=True)
df.drop('is_homework', axis=1, inplace=True)

# Сборка данных по отдельным категориям
df1 = df[df['status'] == 'done']
df2 = df.groupby(['course_title', 'module_number', 'lesson_number']).agg('sum').reset_index()

# Создание общего датафрейма со столбцами студентов которые начали и которые закончили
df = pd.merge(df1, df2, left_on=['course_title', 'module_number', 'lesson_number'], right_on=['course_title', 'module_number', 'lesson_number'], how='inner')
df.rename(columns={'progress_id_x': 'done', 'progress_id_y': 'start'}, inplace=True)
df.drop(columns='status', inplace=True)

# Выбор первого и последнего урока с домашним заданием
g = df.groupby(['course_title', 'module_number'])

df = (pd.concat([g.head(1), g.tail(1)])
   .drop_duplicates()
   .sort_values(['course_title', 'module_number'])
   .reset_index(drop=True))

# Перенос строки выполненных заданий на 1 строку вниз
df['done'] = df['done'].shift(periods=1)

# Замена значений начала курса на нули
g = df.groupby(['course_title'])
df['done'] = np.where(df['start'].isin(g.head(1)['start']), 0, df['done'])

# Перевод столбца в числовое значение и вычисление конверсии
df['done'] = df['done'].astype('int')
df['conversion'] = df['start'] / df['done']

df
```
